```python
import pandas as pd
```

```python
df = pd.read_csv("bitcoin_tweet_2022.csv")
df
```

```python
from collections import defaultdict

def map_reduce(texts):
    word_count = defaultdict(int)
    for text in texts:
        if pd.notnull(text):
            words = text.split()
            for word in words:
                word_count[word] += 1
    sorted_word_count = dict(sorted(word_count.items(), key=lambda item: item[1], reverse=True))
    return sorted_word_count
word_frequencies = map_reduce(df['text2'])
```

```python
top_50_formatted = [
    { "text": word, "value": count }
    for word, count in list(word_frequencies.items())[:50]
]
```

```python
top_50_formatted
```

```python
import json

json_output = json.dumps(top_50_formatted, indent=2, ensure_ascii=False)
with open("../client/src/assets/max_values.json", "w", encoding="utf-8") as f:
    f.write(json_output)
```
