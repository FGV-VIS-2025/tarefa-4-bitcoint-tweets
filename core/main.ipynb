```python
df = cudf.read_csv("Bitcoin_tweets.csv")
```

```python
df = df.dropna()
df = df.drop(columns=["user_description", "is_retweet"])
df
```

```python
def chinese_words(text):
    if pd.isnull(text):
        return False
    return bool(re.search(r'[\u4e00-\u9fff]', text))

pdf = df.to_pandas()

pdf = pdf[
    ~pdf['user_name'].apply(chinese_words) &
    ~pdf['user_location'].apply(chinese_words) &
    ~pdf['text'].apply(chinese_words)
]

df = cudf.DataFrame.from_pandas(pdf)
df = df.reset_index(drop=True)
df
```

```python
def replace_country(location, country_abr):
    for coun, abrev in country_abr:
        if coun.lower() in location.lower() or abrev.lower() in location.lower():
            return coun
    return None 

df['user_location'] = df['user_location'].apply(lambda x: replace_country(x, country_abr))

df = df[df['user_location'].notna()]
df
```

```python
df['hashtags'] = df['hashtags'].apply(lambda x: [hashtag.lower() for hashtag in x] if isinstance(x, list) else (x.lower() if isinstance(x, str) else x))
df['source'] = df['source'].apply(lambda x: [hashtag.lower() for hashtag in x] if isinstance(x, list) else (x.lower() if isinstance(x, str) else x))
df
```

```python
value_counts = df['source'].value_counts()
threshold = value_counts.sum()/150
threshold
```

```python

sources_to_remove = value_counts[value_counts <= threshold].index
df2 = df[~df['source'].isin(sources_to_remove)]
print(df2['source'].value_counts())
print(len(df2['source'].value_counts()))
df2
```
