```python
df['text'] = df['text'].str.replace(r'http\S+|www\S+|https\S+', '', regex=True, case=False)
df['text'] = df['text'].str.replace(r'@\w+', '', regex=True, case=False)
df
```

```python
import unicodedata

def contains_arabic(text):
    arabic_range = re.compile('[\u0600-\u06FF\u0750-\u077F\u08A0-\u08FF]')
    return bool(arabic_range.search(text))

def contains_emoji(text):
    emoji_range = re.compile('[\U0001F600-\U0001F64F\u2600-\u26FF\u2700-\u27BF\u2300-\u23FF\u2B50\u2764\u2763]')
    return bool(emoji_range.search(text))

def clean_dataset(text):
    if pd.isnull(text):
        return text
    text = re.sub(r'\[\s*U\+\w+\s*\]', '', text)
    text = unicodedata.normalize('NFKC', text)

    replace = {
        '‚Äô': "'",
        '‚Äò': "'",
        '‚Äú': '"',
        '‚Äù': '"',
        '¬¥': "'",  
        '`': "'",
    }
    for rare, normal in replace.items():
        text = text.replace(rare, normal)

    text = ''.join(c for c in text if unicodedata.category(c)[0] != 'C')

    text = re.sub(r'\s+', ' ', text)
    text = text.strip()
    
    if contains_arabic(text):
        return None

    return text

def clean_text(text):
    if pd.isnull(text):
        return text

    text = re.sub(r'[?!¬°"#$%&/()=;,_\-\[{+\]*\d]+', '', text)
    text = text.lower()
    text = re.sub(r'[^\w\s]', '', text)
    text = ''.join(c for c in text if unicodedata.category(c)[0] != 'C')
    text = re.sub(r'\s+', ' ', text).strip()
    if not text:
        return None
    
    return text

df['text'] = df['text'].apply(clean_dataset)
df['user_name'] = df['user_name'].apply(clean_dataset)
df['hashtags'] = df['hashtags'].apply(clean_dataset)
df = df.dropna()

df['text2'] = df['text'].apply(clean_text)
```

```python
# Reordenar las columnas para que 'text2' est√© al costado de 'text'
df = df[['user_name', 'user_location', 'user_created', 'user_followers', 'user_friends', 
         'user_favourites', 'user_verified', 'date', 'text', 'text2', 'hashtags', 'source']]
```
